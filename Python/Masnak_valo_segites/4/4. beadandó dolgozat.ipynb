```python
#1.feladat: Hívja be az Értékesítés(nagyker) Excel munkafüzet, E1 és E2 tábláját. Fűzze össze a két
# táblát úgy, hogy csak a közös oszlopok szerepeljenek benne és duplikáció mentes legyen.
```

```python
import pandas as pd
df1 = pd.read_excel('Értekesites(nagyker).xlsx', sheet_name='E1')
df2 = pd.read_excel('Értekesites(nagyker).xlsx', sheet_name='E2', header=None, names=["Dátum", "Kategórianév", "Terméknév", "Mennyiség", "Bevétel", "Költség", "Nyereség"])
common_columns = list(set(df1.columns) & set(df2.columns))
df = pd.merge(df1[common_columns], df2[common_columns], how='outer').drop_duplicates()
df
```

```python
#2.feladat: Fűzze össze az E1 és E3 táblákat is pd.concat()-tal és pd.merge()-el is.
```

```python
df3 = pd.read_excel('Értekesites(nagyker).xlsx', sheet_name='E3')
con_df = pd.concat([df1, df3], ignore_index=True)
con_df
```

```python
mer_df = pd.merge(df1, df3, how='outer')
mer_df
```

```python
#3.feladat: Az E és az E4 tábla segítségével készítsen egy táblát, amiben az E táblát kibővíti a Vezető nevével.
```

```python
dfe = pd.read_excel('Értekesites(nagyker).xlsx', sheet_name='E')
df4 = pd.read_excel('Értekesites(nagyker).xlsx', sheet_name='E4')
merged_df = pd.merge(dfe, df4[['Vevőnév', 'Vezető']], on='Vevőnév', how='left')
merged_df
```

```python
#4.feladat: Az E és az E4 tábla segítségével írja ki azokat a cégeket amelyeknek nem volt még vásárlása.
```

```python
merged_df = pd.merge(df4, dfe, on='Vevőnév', how='left', indicator=True)
```

```python
missing_purchases = merged_df.loc[merged_df['_merge'] == 'left_only', 'Vevőnév'].unique()
```

```python
print("Azok a cégek, amelyeknek még nem volt vásárlása:")
for company in missing_purchases:
    print(company)
```

```python
#5.feladat: Az E tábla segítségével készítsen egy pivot táblát, ahol a sor a Vevőket, az oszlop a
# Kategóriát, az érték pedig az összes Nyereséget mutatja.
```

```python
pivot_table = pd.pivot_table(dfe, values='Nyereség', index='Vevőnév', columns='Kategórianév', aggfunc='sum')
pivot_table
```

```python
#6.feladat: Hívja be a K1 kimutatást, amiből készítsen egy olyan táblát, aminek 3 oszlopa van 
# (Vevő,Város, Költség). A Végösszeg sorokat törölje.
```

```python
#7.feladat: Az E tábla alapján írja ki Vevőnként (azon belül Bolt megyénként) és az oszlopban
# Kategóriánként, hogy hány vásárlás történt.
```

```python
dfe_vasarlasok_szama = dfe.groupby(['Vevőnév', 'Bolt megye', 'Kategórianév']).size().unstack(fill_value=0)
print(dfe_vasarlasok_szama)
```

```python
#8.feladat: Az E tábla alapján számítsa ki, hogy Bolt városonként és Termékenként mennyi az átlagos Nyereség.
```

```python
dfe_atlagos_nyereseg = dfe.groupby(['Bolt város', 'Terméknév'])['Nyereség'].mean().reset_index()
print(dfe_atlagos_nyereseg)
```
