```python
import pandas as pd

#1. Hívja be az Ertekesites (ETL).csv munkafüzetet, a dátumra figyeljen.
df=pd.read_excel('Ertekesites(ETL).xlsx', parse_dates=['Dátum'])
type(df.Dátum[0])
```

```python
#2. Az Egységár és a Érték oszlopot alakítsa egész típusúvá.
df.fillna(0, inplace=True)
df['Egységár'] = df['Egységár'].astype(int)
df['Érték'] = df['Érték'].astype(int)
print(df.dtypes)
```

```python
#3. Egyes oszlopokban hiányzó értékeket talál. Tegye velük ebben sorrendben a következőket:
# a) Ahol nincs érték a Mennyiségben, azt a sor törölje ki.
df1=df.dropna(subset=['Mennyiség'],)
df1
```

```python
# b) A Kategórianév oszlopba kerüljön az „ismeretlen” szó.
df2=df
df2['Kategórianév'].fillna('ismeretlen', inplace=True)
```

```python
# c) Az Egységár hiányzó értékeit töltse ki a terméknév segítségével.
df3=df
df3['Egységár'].fillna(df['Terméknév'], inplace=True)

#ellenőrzése a c feladatrésznek:
missing_values = df[df['Egységár'].isna()]
for index, row in missing_values.iterrows():
    product_name = row['Terméknév']
    print(f"Hiányzó Egységár érték a(z) '{product_name}' termékre")
```

```python
# d) A Érték oszlop legyen az Egységár és a Mennyiség szorzata. Majd nevezze át Bevételnek.
df4=df
df4['Bevétel'] = df['Egységár'] * df['Mennyiség']
df4=df.rename(columns={'Érték':'Bevétel'})
df4
```

```python
#4. A Terméknév mezőben elütések vannak. Korrigálja őket
df5=df
df5.Terméknév=df5.Terméknév.replace(['Tol','Tolll'],'Toll')
df5
df6=df
df6.Terméknév=df6.Terméknév.replace(['creuza','cerza'],'Ceruza')
df6
```

```python
#5. A Mennyiség mezőben vannak elütések:
# a) Ahol negatív értékek vannak, ott legyen a Mennyiség az abszolút érték.
df7=df
df7['Mennyiség'] = df7['Mennyiség'].abs()
print(df['Mennyiség'].head())
```

```python
# b) b) Ahol 200-nál nagyobb Mennyiség van, ott legyen az előző és a következő rekord átlaga.
df8=df
for i in range(len(df8)):
    if df8.loc[i, 'Mennyiség'] > 200:
        
        previous_value = df8.loc[i - 1, 'Mennyiség'] if i > 0 else None
        next_value = df8.loc[i + 1, 'Mennyiség'] if i < len(df8) - 1 else None
        
        average = (previous_value + next_value) / 2 if previous_value is not None and next_value is not None else None
        
        if average is not None:
            df8.loc[i, 'Mennyiség'] = average

print(df8.loc[:10, 'Mennyiség'])
```

```python
#6. Írassa ki az össze duplikációt. Majd szüntesse meg őket úgy, hogy az utolsó sor maradjon csak meg.
duplicates = df[df.duplicated(keep=False)]
df.drop_duplicates(keep='last', inplace=True)
print(df.tail())
```

```python
#7. Mentsen dataframe-t Javított.csv néven, ahol az elválasztó karakter a vessző.
df.to_csv('Javított.csv',index=False)
```
