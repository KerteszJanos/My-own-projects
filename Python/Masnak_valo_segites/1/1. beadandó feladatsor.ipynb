```python
import random

def getRandList(a, f, t):
    nums = set()
    while len(nums) < a:
        nums.add(random.randint(f,t))
    return list(nums)

arr1 = getRandList(1000, 0, 100000)
arr2 = getRandList(1000, 100000, 200000)

concatedArr = arr1 + arr2

rndDir = {}

for i in range(len(arr1)):
    rndDir[arr1[i]] = arr2[i]
    
for i in range(len(arr1)):
    print(f"rndDir[{arr1[i]}] = {arr2[i]}")
```
